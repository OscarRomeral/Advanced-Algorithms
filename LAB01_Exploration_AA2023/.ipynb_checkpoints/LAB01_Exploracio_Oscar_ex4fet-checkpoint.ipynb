- ```Maze(x_num, y_num, seed=None, p_down=0.2, p_build=0)```: Aquesta funció construeix i retorna un laberint aleatori de dimensió $x_\text{num} \times y_\text{num}$. És interessant que, un cop hagueu programat la solució dels exercicis, proveu de modificar-lo per veure que funciona en diferents casos. Així, si $p_\text{down}$ és proper a 1, s'eliminen gairebé totes les parets del laberint, creant molts camins entre totes les cel·les. En canvi, si $p_\text{build}$ és proper a 1, es crearan parets que faran que el laberint no tingui solució per a gairebé cap parella de cel·les.

- ```maze.set_path(path)```: Donat un camí definit com una sequència de nodes, el podem assignar al laberint usant aquesta funció. Recorda que si vols visualitzar un camí assignat, pots fer-ho cridant a ```display_maze(plot_path=True)```.

<h1>Problema 1: BFS (2,5p)</h1>
<p>
    Implementeu l'algorisme <b>BFS</b>. Donat un graf de la llibreria networX, la funció <b>bfs()</b> ha de retornar les caselles accessibles des del paràmetre <b>origin</b>, incloent-hi aquesta, en l'ordre d'exploració. És a dir, la llista retornada sempre inclourà, com a mínim, el node orígen.<br>
    Recordeu que és necessari <b>no canviar l'estructura de la funció</b>. No canvieu ni el nom de la funció, ni el nom dels paràmetres d'entrada ni els de sortida. <br>
</p>    
    
</div>

````python
from collections import deque
def bfs(G, origin):
    """
    Params
    ======    
    :G: Graf. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node orígen
    
    Returns
    =======
    :visited: El conjunt de nodes visitats durant l'exploració BFS en l'ordre en que han estat explorats
    """
    
    #En el cas de recorrer el graf en amplitud és millor fer-ho amb un algoritme iteratiu, en comptes d'un recursiu. 
    
    visited = []
    queue = deque([origin])
    
    while queue:
        node = queue.popleft()  # Agafo el primer node de la llista (Vull fer tots els veïns primer BFS) i l'eliminem de la cua
        
        if node not in visited: # Comprovem si ja està visitat
            print(node)
            visited.append(node)
            
        for neighbor in G.neighbors(node): #guardem els veins del node al final de la cua 
                if neighbor not in visited:
                    queue.append(neighbor) # Com en cada iteració del while agafem el primer node, estarem recorrent per amplada 
    
    return visited

# Prova del funcionament de la funció bfs
G = nx.Graph()
G.add_edges_from([('1','2'),('2','3'),('3','4'),('1','5'),('5','6'),('6','7'),('7','8'),('4','8'),('1','6'),('2','8')]) # Graph d'exemple
print(bfs(G, '1'))
````

````python
# Obtenim el graph associat al laberint on cada node es una cel·la i tindrem una aresta si entre dues cel·les no existeix paret
maze = Maze(7, 7, p_build=0.2, seed=8)
G = maze.maze_graph()

A = 24                    
visited = bfs(G, A)
num = len(visited)

print(f"Des de la cel·la {A} podem accedir a {num} caselles")
print(f"Ordre d'exploració: {visited}")
maze.set_values(visited, 1, color=(255,0,0))
maze.set_value(A, 1, color=(0,255,0))

maze.display_maze(plot_path=True, height=200)
````

<h1>Problema 2: Holes (2,5p)</h1>
<p>
    Implementeu una funció anomenada <b>holes()</b> que, donat un graf, un node orígen, un node destí, i un diccionari de nodes amb penalitzacions, trobi el camí més curt entre orígen i destí.<br>
    Si el camí passa per un dels nodes del diccionari, el cost de visitar el node serà la penalització en comptes de ser cost 1.<br>
    Heu de basar la vosta solució en l'algorisme <b>Dijkstra</b>.
</p>    
    
</div>

````python
import heapq # Utilitzarem la cua de prioritat en l'algoritme
def holes(G, origin, destination, holes_dct={}):
    """
    Params
    ======
    :G: Graf del qual en volem extreure el camí mínim. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node orígen
    :destination: Índex del node destí
    :holes_dct: Un diccionari del tipus {node: penalització}. Passar per cada node del diccionari té un cost diferent a 1.
    
    Returns
    =======
    :path: Una llista de nodes del camí més curt entre els nodes 'origin' i 'destination' (ambdós inclosos).
    :cost: Un enter amb el cost de recórrer el camí, incloent-hi les penalitzacions.
    """
    
    path = []
    cost = 0
    infinity = 2**32-1 # Nombre infinit gran
    
    if destination == origin:
        path = [origin]
        return path, cost
    else: # Cas no trivial
        heap = []
        

        # Farem ús de diccionaris del tipus: {node: [dist, prev]}, dist = cost actual prev = node previ al actual
        visitat = dict()
        informacio = dict()
        heapq.heappush(heap, (0, origin)) # Posem a la cua de prioritat el primer element 
        for node in G.nodes():
            informacio.update({node: [infinity, None]}) # Informació conté els nodes del graph amb el seu cost inicialitzat a infinit i sense node previ.
            heapq.heappush(heap, (infinity, node)) # Notem que heapq conté tots els nodes del graph
            visitat.update({node: False}) # Diccionari dels nodes visitats 
        informacio.update({origin: [0, None]})

        trobat = False # Farem un bucle d'iteracions fins que el node actual sigui el destination
        while(len(heap) != 0 and not trobat):
            nodeActual = heapq.heappop(heap)[1] # Treu i ens retorna el element més petit de heapq, l'índex [1] l'utilitzem per obtenir el node 
            if not visitat[nodeActual]:
                visitat.update({nodeActual: True})
                for neighbor in G.neighbors(nodeActual): # Pels veïns del nodeActual calculem la distancia alternativa del algortime de Dijkstra
                    if neighbor in holes_dct:
                        distancia_alternativa = informacio[nodeActual][0] + holes_dct[neighbor]
                    else:
                        distancia_alternativa = informacio[nodeActual][0] + 1
                    if (distancia_alternativa < informacio[neighbor][0]):
                        informacio.update({neighbor: [distancia_alternativa, nodeActual]}) # Actualitzem el cost del nodeActual
                        heapq.heappush(heap, (distancia_alternativa, neighbor)) # Posem el veï a la cua de heapq
            trobat = nodeActual == destination
       
        if(not trobat):
            cost = -1
        else:
            #Creo el path
            #Iteraré fins arribar al node d'origen, i l'afegiré al path    
            busco = destination
            fi = False
            while not fi:
                path.append(busco)
                aux = informacio[busco][1]
                if busco in holes_dct:
                    cost += holes_dct[busco]
                else:
                    cost += 1
                busco = aux
                fi = (origin == busco)
            path.append(busco)

            #Inverteixo el path ja que he estat fent servir el mètode append, que afegeix els elements al final, perquè té menys cost computacional
            path = path[::-1]
    
    return path, cost
````

````python
# Proveu el vostre algorisme!

maze = Maze(30, 10, seed=15)

# Obtenim el graph associat al laberint on cada node es una cel·la i tindrem una aresta si entre dues cel·les no existeix paret
G = maze.maze_graph()

A = 0
B = len(G.nodes)-1
color = (255, 0, 255)

h_dcts = [
    {}, # Sense penalització
    {8:20,72:2, 227:10},
    {8:15, 72:10, 227:9},
]
for holes_dct in h_dcts:
    maze.set_values(list(holes_dct.keys()), 1, color)
    path, cost = holes(G, A, B, holes_dct)

    maze.set_path(path)
    print(f"Cost total: {cost}")
    print('Longitud del camí:', len(path))
    maze.display_maze(plot_path=True, height=200)

    maze.set_values(list(holes_dct.keys()), 0, color)
````

<h1>Problema 3: Checkpoint (2,5p)</h1>
<p>
    Implementeu una funció anomenada <b>checkpoint()</b> que, donat un graf, un node origen, un node destí i un node extra, trobi el camí més curt des del node origen fins el node destí passant pel node extra. Com en l'exercici anterior, tindrem un diccionari de punts on aplicarem una penalització.<br>
</p>    
    
</div>

<div class="alert alert-warning">
<p>
    Quina és la complexitat d'aquest algorisme?</b><br>
</p>    

 << La complexitat de l'algoritme depèn del nombre de parades que fem. A cada parada fem una crida a la funció Holes (que essencialment és un Dijkstra que té com bé sabem per teoria E*log V, on E = edges V = vèrtexs). En definitiva, si fem n parades, l'algoritme tindrà complexitat n*E*log V . En el nostre cas, n = 2, podem dir doncs que és la mateixa que la del Dijkstra.>>
    
</div>

````python
import heapq

def checkpoint(G, origin, destination, extra, holes_dct={}):
    """
    Params
    ======
    :G: Grafo del cual queremos extraer el camino mínimo. Debe ser un objeto de la clase nx.Graph.
    :origin: Índice del nodo de origen.
    :destination: Índice del nodo de destino.
    :extra: Índice del nodo extra por el que debe pasar el camino.
    :holes_dct: Un diccionario del tipo {nodo: penalización}. Pasar por cada nodo del diccionario tiene un costo diferente a 1.

    Returns
    =======
    :path: Una lista de nodos del camino más corto entre los nodos 'origin' y 'destination' (ambos incluidos).
    :cost: Un entero con el costo de recorrer el camino, incluyendo las penalizaciones.
    """

    path = []
    cost = 0
    infinity = 2**32-1
    trobat = False
    path, cost = holes(G, origin, extra, holes_dct)
    if cost != -1:
        #S'ha trobat un camí de extra a destination
        path2, cost2 = holes(G, extra, destination, holes_dct)
        if path2 and cost2 != -1:
                # S'ha trobat un camí de extra a destination
                path += path2[1:]  # Excluim el primer node per no duplicar
                cost += cost2
        
    return path, cost
````

````python
# Proveu el vostre algorisme!

# Creem un laberint i n'extraiem el graph
maze = Maze(6,6, seed=17)
G = maze.maze_graph()

# Definim l'inici, el fi, un node extra i la llista amb penalització
A = 30
B = len(G.nodes)-1
E = 8

h_dcts = [
    {}, # Sense penalització
    {13:3, 28:5},
    {13:10, 28:20},
]
for holes_dct in h_dcts:
    # La funció set_value ens permet 'activar' un node per a que surti dibuixat al laberint.
    maze.set_value(E, 1)                         # Checkpoint
    maze.set_values(list(holes_dct.keys()), 1, (255,0,255))  # Penalitzacions

    path, cost = checkpoint(G, A, B, E, holes_dct)
    print(f'El cost del camí més curt que passa per E és: {cost}')
    print('La distàcia del camí més curt que passa per E és:', len(path))
    print(f'El camí és: {path}')
    maze.set_path(path)
    maze.display_maze(plot_path=True, height=200)
    
    maze.set_value(E, 0)                         # Checkpoint
    maze.set_values(list(holes_dct.keys()), 0, (255,0,255))  # Penalitzacions
    
````

<h1>Problema 4: Checkpoints list (2,5p)</h1>
<p>
    Implementeu una funció anomenada <b>checkpoints_list()</b> que, donat un graf, un node origen, un node destí i una llista de nodes extres, trobi el camí més curt des del node origen fins el node destí passant per tots els nodes extra, en l'ordre més òptim. Com en l'exercici anterior, tindrem un diccionari de punts on aplicarem una penalització.<br>
    <b>Ajuda</b>: Podeu fer servir la funció <b>itertools.permutations</b>.
</p>    
    
</div>

<div class="alert alert-warning">
<p>
    Quina és la complexitat d'aquest algorisme?</b><br>
</p>    

 << La complexitat de l'algoritme depèn del nombre de parades que fem. A cada parada fem una crida a la funció Holes (que essencialment és un Dijkstra que té com bé sabem per teoria E*log V, on E = edges V = vèrtexs). En definitiva, si fem n parades, l'algoritme tindrà complexitat n*E*log V . En el nostre cas, n = 2, podem dir doncs que és la mateixa que la del Dijkstra >>
    
</div>

````python
from itertools import permutations
def checkpoints_list(G, origin, destination, extras, holes_dct={}):
    """
    Params
    ======
    :G: Graf del qual en volem extreure el camí mínim. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node orígen
    :destination: Índex del node destí
    :extras: Llista d'índexs de nodes per on ha de passar el camí.
    :holes_dct: Un diccionari del tipus {node: penalització}. Passar per cada node del diccionari té un cost diferent a 1.
    
    Returns
    =======
    :path: Una llista de nodes del camí més curt entre els nodes 'origin' i 'destination' que passa per tots els nodes 'extras'.
    :cost: Un enter amb el cost de recórrer el camí, incloent-hi les penalitzacions.
    """
    
    path = []
    cost = 0
    
    all_nodes = [origin] + extras + [destination]
    parelles_repetides = list(permutations(all_nodes, 2))
    parelles = []
    for parella in parelles_repetides:
        if parella not in parelles and parella[::-1] not in parelles: #D'aquesta manera no tenim parelles de nodes repetits pero en ordre invers
            parelles.append(parella) #Ens estalvia la meitat dels dijkstres que calcularem posteriorment
            
    parella_dijkstra = dict()
    pair_path = []
    pair_cost = 0
    for parella in parelles: #Per cada parella de nodes, calculem la ruta mínima amb dijkstra i la guardem en un diccionari amb la parella com a clau
        pair_path , pair_cost = holes(G, parella[0], parella[1], holes_dct)
        parella_dijkstra[parella] = [pair_path, pair_cost]
    
    lenght_extras = len(extras)
    extras_combinations = list(permutations(extras, lenght_extras)) #Fem les permutacions possibles entre els nodes del mig ja que origin i destination no canvien
    general_combinations = []
    for i in range(len(extras_combinations)):
        general_combinations.append((origin,) + extras_combinations[i] + (destination,)) #Afegim, per cada combinacio de extres, el origen i final 
        
    final_path = []
    final_cost = 2**32-1
    for combinacio in general_combinations:
        combination_path = [origin]
        combination_cost = 0
        print(combinacio)
        for i in range(len(combinacio)-1): #Per cada combinacio calcularem el path total i el seu cost per tal de veure quin es menys costos
                if (combinacio[i], combinacio[i+1]) in parelles: #Comprovem aquest if ja que els recorreguts entre dos nodes nomes els tenim guardats per parells de nodes en un sentit, es a dir, tenim (10,27) pero no (27,10) i es el mateix camí i cost
                    combination_path += parella_dijkstra[(combinacio[i], combinacio[i+1])][0][1:]
                    combination_cost += parella_dijkstra[(combinacio[i], combinacio[i+1])][1]
                else: #En cas de no tenir la parella com a key en el diccionari de recorreguts, agafem la key invertint la parella
                    reversed_path = parella_dijkstra[(combinacio[i+1], combinacio[i])][0][1:][::-1] #Com el recorregut està guardat en sentit contrari, guardem el recorregut invertint-lo amb [::-1]
                    combination_path += reversed_path
                    combination_cost += parella_dijkstra[(combinacio[i+1], combinacio[i])][1]
        if final_cost > combination_cost: #Aqui guardem com a path final el recorregut més curt en quant a cost per a totes les combinacions possibles
            final_path, final_cost = combination_path, combination_cost
            print("\nCombi final: ")
            print(combinacio)
            print("\n")
        
    
    return final_path, final_cost
````

````python
# Proveu el vostre algorisme!

# Creem un laberint i n'extraiem el graph
maze = Maze(30, 10, seed=25)
G = maze.maze_graph()

# Definim l'inici, el fi i node extres
A = 20
B = len(G.nodes)-1
E = [10,187, 286, 27]

h_dcts = [
    {}, # Sense penalització
    {216:200, 208:30, 84:50},
    {79:500, 127:100, 84:5, 15: 500}
]
for holes_dct in h_dcts:
    # La funció set_values ens permet 'activar' nodes per a que surtin dibuixat al laberint.
    maze.set_values(E, 1)                        # Checkpoint
    maze.set_values(list(holes_dct.keys()), 1, (255,0,255))  # Penalitzacions

    path, cost = checkpoints_list(G, A, B, E, holes_dct)
    print(f'El cost del camí més curt que passa per E és: {cost}')
    print('La distàcia del camí més curt que passa per E és:', len(path))
    maze.set_path(path)
    maze.display_maze(plot_path=True, height=200)

    # De la mateixa forma la funció set_values també ens permet 'desactivar' els nodes.
    maze.set_values(E, 0)
    maze.set_values(list(holes_dct.keys()), 0, (255,0,255))
    
````
